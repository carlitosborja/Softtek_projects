## Task 1 - Detecting changes

```python
# First we import some necessary libraries:
import numpy as np
import pandas as pd

# We import the created module as pp
import Python_pandas_project as pp
```

```python
# Creation of the first DataFrame woth the dates and the weather

date = ['1/' + str(i) + '/20' for i in range(1,11)]
was_rainy = [False, True, True, False, False, True, False, True, True, True]

data_rainy = {'date': date, 'was_rainy': was_rainy}

df_rainy = pd.DataFrame(data_rainy)
```

## Task 2 - Seasons problem

```python
# We create the pandas DataFrame with the sample information

ORD_ID = ['113-8909896-6940269', '114-0291773-7262677', '114-0291773-7262697', 
          '114-9900513-7761000', '112-5230502-8173028', '112-7714081-3300254', 
          '114-5384551-1465853', '114-7232801-4607440']
ORD_DT = ['9/23/19', '1/1/20', '12/5/19', '9/24/20', '1/30/20', '5/2/20', '4/2/20', '10/9/20']
QT_ORDD = np.ones(8)

data_seasons = {'ORD_ID': ORD_ID, 'ORD_DT': ORD_DT, 'QT_ORDD': QT_ORDD}

df_seasons = pd.DataFrame(data_seasons)

df_seasons
```

```python
# It is necessary to change the data type for the column ORD_DT from string to Timestamp,
# in order to be able to apply the function pp.seasons to this column

df_seasons.ORD_DT = pd.to_datetime(df_seasons.ORD_DT)
df_seasons
```

```python
# Now we just apply the function pp.seasons to map each date in the column ORD_DT to its respective season of the year.

df_seasons['SEASON'] = df_seasons.ORD_DT.apply(pp.seasons)
df_seasons[['ORD_ID','SEASON']]
```
